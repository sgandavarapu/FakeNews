```python
import urllib2
from bs4 import BeautifulSoup

url = 'http://100percentfedup.com'
req = urllib2.Request(url, headers={'User-Agent' : "Magic Browser"}) 
page = urllib2.urlopen( req )
#print page.read()
```

```python
#Parse the html in the 'page' variable, and store it in Beautiful Soup format
soup = BeautifulSoup(page)
print soup
```

```python
all_links = soup.find_all('a')
chosen_articles = []
for link in all_links:
    article = link.get('href')
    if article not in chosen_articles:
        chosen_articles.append(article)
        print article
    else:
        continue
print 'number of unique articles in the website are: %d' % len(chosen_articles)
```

```python

#url = 'http://100percentfedup.com/category/education/'

for url in chosen_articles:
    print url
    try:
        req = urllib2.Request(url, headers={'User-Agent' : "Magic Browser"}) 
        page1 = urllib2.urlopen( req )
        #Parse the html in the 'page' variable, and store it in Beautiful Soup format
        soup1 = BeautifulSoup(page1)
        #print soup1
        all_text = soup1.find_all('p')
        if len(all_text) > 10:
            title = soup1.find_all('title')
            print title

            for text in all_text:
                print text

    except ValueError:
        continue
```
