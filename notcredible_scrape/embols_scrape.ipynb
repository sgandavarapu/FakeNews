```python
import requests
import json
import os
from bs4 import BeautifulSoup
from datetime import datetime
import newspaper
```

```python
source = 'embols'
month = '03'
days = [str(i) for i in list(range(1,10))]

dates = []
for day in days:
    dates.append("{0}-0{1}".format(month, day))
```

```python
for date in dates:
    # Load HTML
    filepath = '/mnt/c/Users/Brennan/Dropbox/MIDS/w210-capstone/data/notCredible/{0}/{1}.html'.format(source,date)
    with open(filepath) as f:
        soup = BeautifulSoup(f, 'lxml')

    # Pull out article links
    links = []
    i=0

    for article in soup.findAll('article'):
        for span in article.findAll('span', attrs={"class": "thetime updated"}):
            for span2 in span.findAll('span'):
                hours = range(1,25)
                hours_str = []
                for hour in hours:
                    hours_str.append(str(hour))
                if span2.text.replace(' hours ago','') in hours_str:
                    for header in article.findAll('h2', attrs={"class":"title front-view-title"}):
                        for link in header.findAll('a'):
                            links.append(link.get('href'))

    links = set(links)
    links = list(links)
    
    for link in links:
        link = newspaper.Article(link)
        link.download()
        html = link.html
        link.parse()
        date_longform = '{0}-2017'.format(date)
        title = link.title
        
        article = {}
        article["html"] = html
        article["title"] = title
        article["url"] = link.url
        article["date"] = date_longform
        article["source"] = source
        article["text"] = link.text
        article["images"] = list(link.images)
        article["videos"] = link.movies

        outfile = '/mnt/c/Users/Brennan/Dropbox/MIDS/w210-capstone/data/notCredible/{0}/articles/{2}/article{1}_{2}.txt'.format(source,i,date)
        os.makedirs(os.path.dirname(outfile), exist_ok=True)
        i+=1

        with open(outfile, 'w') as f:
            json.dump(article, f)
```
