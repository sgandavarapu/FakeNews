```python
import requests
import json
import os
from bs4 import BeautifulSoup
from datetime import datetime
import newspaper
```

```python
source = 'dcclothesline'
```

```python
month = '02'
days = [str(i) for i in list(range(18,29))]

dates = []
for day in days:
    if len(day)<2:
        day = "0{}".format(day)
    dates.append("{0}-{1}".format(month, day))
    
month='03'
days = [str(i) for i in list(range(1,14))]

for day in days:
    if len(day)<2:
        day = "0{}".format(day)
    dates.append("{0}-{1}".format(month, day))
```

```python
for date in dates:
    # Load HTML
    filepath = '/ebs_volume/data/notCredible/{0}/{1}.html'.format(source,date)
    try:
        with open(filepath) as f:
            soup = BeautifulSoup(f, 'lxml')

        # Pull out article links
        links = []
        i=0

        pubdates=[]

        for div in soup.findAll('div', attrs={'id':'content'}):
            for span in div.findAll('span', attrs={'class': 'entry-date'}):
                pubdate = datetime.strptime(span.text, "%B %d, %Y")
                pubdates.append(str(pubdate))
            for header in div.findAll('h2', attrs="entry-title"):
                for link in header.findAll('a'):
                    links.append(link.get('href'))

        links = set(links)
        links = list(links)
        updated_links=[]

        for l, d in zip(links, pubdates):
            if date in str(pubdates):
                updated_links.append(l)

        for link in updated_links:      
            link = newspaper.Article(link)
            link.download()
            html = link.html
            link.parse()
            date_longform = '{0}-2017'.format(date)
            title = link.title.replace(' | ','')

            article = {}
            article["html"] = html
            article["title"] = title
            article["url"] = link.url
            article["date"] = date_longform
            article["source"] = source
            article["text"] = link.text
            article["images"] = list(link.images)
            article["videos"] = link.movies

            outfile = '/ebs_volume/data/notCredible/{0}/articles/{1}/article{2}.txt'.format(source,date,i)
            os.makedirs(os.path.dirname(outfile), exist_ok=True)
            i+=1

            with open(outfile, 'w') as f:
                json.dump(article, f)
                
    except FileNotFoundError:
        print("File not found")
```
