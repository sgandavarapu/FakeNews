```python
import requests
import os
import json
from bs4 import BeautifulSoup
import schedule
import datetime
import newspaper
```

```python
urls = []

with open('/ebs_volume/persistent_scripts/fakeNewsSites.txt') as infile:
    for line in infile:
        url = ('http://' + line).replace('\n','')
        urls.append(url)
```

```python
month = '02'
days = [str(i) for i in list(range(18,29))]

dates = []
for day in days:
    if len(day)<2:
        day = "0{}".format(day)
    dates.append("{0}-{1}".format(month, day))
    
month='03'
days = [str(i) for i in list(range(1,14))]

for day in days:
    if len(day)<2:
        day = "0{}".format(day)
    dates.append("{0}-{1}".format(month, day))
```

```python
for date in dates:
    for url in urls:
        # Load HTML
        source = url.replace('http://','').replace('.com','').replace('.org', '').replace('.us', '').replace('.me', '').replace('.net', '')
        print('--{}...'.format(source))
        filepath = '/ebs_volume/data/notCredible/{0}/{1}.html'.format(source,date)
        try:
            with open(filepath) as f:
                soup = BeautifulSoup(f, 'lxml')

            # Pull out article links that were published on date
            links = []
            i=0
            
            for html_link in soup.find_all('a', {'rel': 'bookmark'}):
                url = html_link.get('href')
                if url not in links:
                    links.append(url)

            for link in links:
                try:
                    link = newspaper.Article(link)
                    link.download()
                    html = link.html
                    link.parse()
                    pubdate = str(link.publish_date)
                    if date in pubdate:
                        date_longform = '{0}-2017'.format(date)
                        title = link.title
                        article = {}
                        article["html"] = html
                        article["title"] = title
                        article["url"] = link.url
                        article["date"] = date_longform
                        article["source"] = source
                        article["text"] = link.text
                        article["images"] = list(link.images)
                        article["videos"] = link.movies

                        outfile = '/ebs_volume/data/notCredible/{0}/articles/{1}/article{2}.txt'.format(source,date,i)
                        os.makedirs(os.path.dirname(outfile), exist_ok=True)
                        i+=1

                        with open(outfile, 'w') as f:
                            json.dump(article, f)
                except:
                    pass
        
        except FileNotFoundError:
            print("File not found")
    
    print("{} Finished".format(date))
    
print("Article Download Complete")
```
